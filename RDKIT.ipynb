### The simplified molecular-input line-entry system (SMILES) is a specification in the form of a line notation for describing the structure of chemical species 

```python
mol=Chem.MolFromSmiles('CN1CC(c2cccc(C3CCC(C(=O)NCc4n[nH]c(-c5ccoc5)n4)CC3)c2)CC1=O')
if mol is not None:
    cs=Chem.MolToSmiles(mol,isomericSmiles=True)
    fingerprint = AllChem.GetMorganFingerprintAsBitVect(mol, 2, nBits=1024)
print(cs)
print(fingerprint)
```

```python
smile=Chem.MolToSmiles(mol)
smile
```
