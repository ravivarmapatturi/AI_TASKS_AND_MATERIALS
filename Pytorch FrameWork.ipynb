## Datasets and Dataloaders in pytorch

#### In PyTorch, Dataset and DataLoader are two important classes for handling data.
torch.utils.data.DataLoader and torch.utils.data.Dataset that allow you to use pre-loaded datasets as well as your own data. Dataset stores the samples and their corresponding labels, and DataLoader wraps an iterable around the Dataset to enable easy access to the samples.



1.Dataset: The Dataset class is an abstract class representing a dataset. Your custom dataset should inherit Dataset and override the following methods

        1:__len__ so that len(dataset) returns the size of the dataset.
        __getitem__ to support the indexing such that dataset[i] can be used to get the ith sample.
        The Dataset stores the samples and their corresponding labels
        2.PyTorch domain libraries provide a number of pre-loaded datasets (such as FashionMNIST) that subclass torch.utils.data.Dataset and implement functions specific to the particular data

2.DataLoader: The DataLoader class wraps an iterable around the Dataset to enable easy access to the samples. It also provides various utilities like batching, shuffling the data, and loading the data in parallel using multiprocessing workers.

### how to create custom dataset in pytorch

```python
class CustomDataset(Dataset):
    def __init__(self,features,labels):
        #data loading 
        self.features=features
        self.labels=labels
        
    def __getitem__(self,idx):
        #to access samples with index
        features=self.features[idx]
        labels=self.labels[idx]
        return features,labels
    
    def __len__(self):
        #to find the length of the data samples 
        return len(self.features)
        
```

```python
df=pd.DataFrame({"features":np.arange(1,20,2),"labels":np.arange(1,40,4)})
```

```python
df
```

```python
dataset=CustomDataset(df.features,df.labels)

dataloader=DataLoader(dataset=dataset,batch_size=5,shuffle=True)
```

```python
for x,y in dataloader:
    print(x,y)
```

### Models building in pytorch

### Linear Regression

```python
import numpy as np
import torch.nn as nn
import torch
import matplotlib.pyplot as plt
from sklearn import datasets
import seaborn as sns

x_numpy,y_numpy=datasets.make_regression(n_samples=100, n_features=1, noise=20, shuffle=True,random_state=23)
```

```python
plt.plot(x_numpy,y_numpy,'ro')
plt.xlabel("xaxis")
plt.ylabel("yaxis")
plt.title("data")
```

```python
x=torch.tensor(x_numpy,dtype=torch.float32)
y=torch.tensor(y_numpy,dtype=torch.float32)
y=y.view(y.shape[0],1)
```

```python
x.shape,y.shape,x.dtype
```

```python
input_size=1
output_size=1
```

```python
model=nn.Linear(input_size,output_size)
```

```python
for param in model.parameters():
    print(param)
model_state_dict = model.state_dict()
weights = model_state_dict['weight']
biases = model_state_dict['bias']

print("Weights:")
print(weights)

print("\nBiases:")
print(biases)
```

```python
## criterion and optimizer
```

```python
criterion=nn.MSELoss()
optimizer=torch.optim.SGD(model.parameters(),lr=0.01)
```

```python
n_epochs=1000
for i in range(n_epochs):
    y_pred=model(x)
    loss=criterion(y_pred,y)
    
    loss.backward()
    optimizer.step()
    optimizer.zero_grad()
    
    if i%100==0:
        print("epoch: {}  loss: {} weight :{} ".format(i,loss.item(),model_state_dict['weight']))
```

```python
pred=model(x).detach().numpy()
plt.plot(x_numpy,y_numpy,'ro')
plt.plot(x_numpy,pred,'b')
plt.xlabel("xaxis")
plt.ylabel("yaxis")
plt.title("prediction")
```
